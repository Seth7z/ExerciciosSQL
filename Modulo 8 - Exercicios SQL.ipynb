A) Listar nome e sobrenome ordenado por sobrenome

```python
cursor.execute("SELECT PrimeiroNome, SegundoNome FROM Funcionarios ORDER BY SegundoNome")
print(cursor.fetchall())
```

B) Listar todos os campos de funcionários ordenados por cidade

```python
cursor.execute("SELECT * FROM Funcionarios ORDER BY Cidade")
print(cursor.fetchall())
```

C) Liste os funcionários que têm salário superior a R$ 1.000,00 ordenados pelo nome completo

```python
cursor.execute("SELECT PrimeiroNome, SegundoNome, UltimoNome FROM Funcionarios WHERE Salario > 1000.00")
print(cursor.fetchall())
```

D) Liste a data de nascimento e o primeiro nome dos funcionários ordenados do mais novo para o mais velho

```python
cursor.execute("SELECT DataNasci, PrimeiroNome FROM Funcionarios ORDER BY strftime('%TY-%m-%d', DataNasci) DESC")
print(cursor.fetchall())
```

E) Liste o total da folha de pagamento

```python
cursor.execute("SELECT SUM(Salario) FROM Funcionarios")
print(cursor.fetchall())
```

F) Liste o nome, o nome do departamento e a função de todos os funcionários

```python
sql = """SELECT
         Funcionarios.PrimeiroNome,
         Departamentos.Nome,
         Funcionarios.Funcao
         FROM Departamentos
         INNER JOIN Funcionarios ON Funcionarios.Codigo = Departamentos.Codigo"""
cursor.execute(sql)
cursor.fetchall()
```

G) Liste a quantidade de funcionários desta empresa

```python
cursor.execute("SELECT COUNT(Codigo) FROM Funcionarios")
cursor.fetchall()
```
